```python
#the root of the project
PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
print(PROJECT_ROOT)
```

```python
# Add the project root to the Python module search path
if PROJECT_ROOT not in sys.path:
    sys.path.insert(0, PROJECT_ROOT)

print(f"Project root configured: {PROJECT_ROOT}")
```

```python
# --- 2. LOADING THE CONFIG.YAML FILE ---

# The config.yaml file is expected to be at the project root.
CONFIG_PATH = os.path.join(PROJECT_ROOT, "config.yaml")
try:
    with open(CONFIG_PATH, 'r') as file:
        config = yaml.safe_load(file)
    print("Configuration loaded successfully.")
except FileNotFoundError:
    print(f"Error: Configuration file not found at {CONFIG_PATH}")
    sys.exit(1)  # Stop execution if the configuration is missing
```

```python
# --- 3. DEFINING PATHS AND AUTOMATICALLY CREATING FOLDERS ---

# Concatenate the project root with the relative paths from the YAML
FIGURES_DIR = os.path.join(PROJECT_ROOT, config['paths']['figures'])
MODELS_DIR = os.path.join(PROJECT_ROOT, config['paths']['models'])
DATA_RAW_DIR = os.path.join(PROJECT_ROOT, config['paths']['data_raw'])

# Automatically create required folders (figures, models, raw data)
os.makedirs(FIGURES_DIR, exist_ok=True)
os.makedirs(MODELS_DIR, exist_ok=True)

print(f"Folders created: {FIGURES_DIR} and {MODELS_DIR}")
```
